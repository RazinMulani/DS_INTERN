## Slicing in pandas
- We can use `iloc` to slice a list based on it's indices.
- `.iloc[row_slice, column_slice]`



```python
x = df.iloc[3:4, 1:6]
x
```

We can also use `iloc` to get specific rows from our dataset.

```python
y = df.iloc[[2,3,4,1], [1,2]]
y
```

Another way of slicing a dataset is using `loc`. This uses labels and not indices to slice.

```python
df.loc[:, 'name']
```

```python
df.loc[:, ['name', 'marks']]
```

```python
df['name']
```

```python
df.name
```

```python
df.iloc[list(range(5)) + list(range(6,10))]
```

`.drop()` method can be used to drop a row or column from our dataset. It takes axis as an argument. <br>

`axis = 1` : col

```python
df.drop(5)
```

```python
df.drop(['name', 'class'], axis = 1)
```
